```python
#first-class function
from mpmath import *
def f(x1, x2):
  return x1**2 + x2, 5*x1**2 - 3*x1 + 2*x2 - 3

r=findroot(f, (0, 0))
print(f'root={r}')
```

```python
#default value of parameters in a function
def printChars(p="*",n=5):
  text=str("")
  print(f'p={p} , n={n}')
  for i in range(1,n+1):
    text=text+p
  return text
```

```python
#Test printChars
text=printChars() #ok
#text=printChars(10) #error
text=printChars("-") #ok
text=printChars("*",10)  #ok
print(text)
```

```python
def no_side_effects(cities):
    print(cities)
    cities = cities + ["Birmingham", "Bradford"]
    print(cities)
locations = ["London", "Leeds", "Glasgow", "Sheffield"]
no_side_effects(locations)
```

```python
def side_effects(cities):
    print(cities)
    cities += ["Birmingham", "Bradford"]
    print(cities)
 
locations = ["London", "Leeds", "Glasgow", "Sheffield"]
```

```python
def side_effects(cities):
    print(cities)
    cities += ["Paris", "Marseille"]
    print(cities)
 
locations = ["Lyon", "Toulouse", "Nice", "Nantes", "Strasbourg"]
side_effects(locations[:])
print(locations) 
```
